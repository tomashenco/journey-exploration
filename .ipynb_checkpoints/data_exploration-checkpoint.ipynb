```python
import pandas as pd
import os

dir_path = 'all journeys'
```

```python
file_names = [file_name for file_name in os.listdir(dir_path) if '.csv' in file_name]
test_file = file_names[0]
data = pd.read_csv(os.path.join(dir_path, test_file))
data.head()
```

```python
data['timestamp'] = pd.to_datetime(data['timestamp'], unit='ms')
data.head()
```

```python
data['type'].unique()
```

```python
data_gps = data.loc[data['type'] == 'gps'].drop(['type', 'x', 'y', 'z'], axis=1)
data_gps.head()
```

```python
data_acc = data.loc[data['type'] == 'accelerometer'].drop(['type', 'lat', 'lon', 'height', 'accuracy', 
                                                           'speed', 'bearing'], axis=1)
data_acc.head()
```
